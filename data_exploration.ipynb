```python
import pandas as pd

df = pd.read_csv("numbeo_raw.csv")

print(df.shape)
print(df.dtypes)
df.head(10)
```

```python
import re

def parse_price(val):
    if pd.isna(val):
        return None
    # Remove currency symbols, letters, spaces — keep digits, dot, comma
    cleaned = re.sub(r'[^\d,\.]', '', str(val))
    # Remove commas used as thousands separators
    cleaned = cleaned.replace(',', '')
    try:
        return float(cleaned)
    except:
        return None

# Apply to all columns except 'city' and the already-numeric mortgage column
price_cols = [c for c in df.columns if c not in ['city', 'Annual Mortgage Interest Rate (20-Year Fixed, in %)']]

df_clean = df.copy()
df_clean[price_cols] = df[price_cols].map(parse_price)

print(df_clean.dtypes)
df_clean.head(3)
```

```python
missing = df_clean.isnull().sum()
missing_pct = (missing / len(df_clean) * 100).round(1)

summary = pd.DataFrame({'missing_count': missing, 'missing_pct': missing_pct})
summary[summary.missing_count > 0].sort_values('missing_pct', ascending=False)
```

```python
# Daily cost formulas per tier (per person)
def calculate_daily_costs(row):
    results = {}

    # --- FOOD ---
    results['food_budget'] = (
        row['Meal at an Inexpensive Restaurant'] * 2 +
        row['Combo Meal at McDonald\'s (or Equivalent Fast-Food Meal)'] * 1
    )
    results['food_mid'] = (
        row['Meal at an Inexpensive Restaurant'] * 1 +
        (row['Meal for Two at a Mid-Range Restaurant (Three Courses, Without Drinks)'] / 2) * 1 +
        row['Combo Meal at McDonald\'s (or Equivalent Fast-Food Meal)'] * 1
    )
    results['food_comfort'] = (
        (row['Meal for Two at a Mid-Range Restaurant (Three Courses, Without Drinks)'] / 2) * 3
    )

    # --- TRANSPORT ---
    results['transport_budget'] = (
        row['One-Way Ticket (Local Transport)'] * 2
    )
    results['transport_mid'] = (
        row['One-Way Ticket (Local Transport)'] * 2 +
        row['Taxi Start (Standard Tariff)'] +
        row['Taxi 1 km (Standard Tariff)'] * 3
    )
    results['transport_comfort'] = (
        row['Taxi Start (Standard Tariff)'] * 3 +
        row['Taxi 1 km (Standard Tariff)'] * 5
    )

    # --- ACCOMMODATION ---
    results['accommodation_budget'] = (
        row['1 Bedroom Apartment Outside of City Centre'] / 30
    )
    results['accommodation_mid'] = (
        (row['1 Bedroom Apartment Outside of City Centre'] +
         row['1 Bedroom Apartment in City Centre']) / 2 / 30
    )
    results['accommodation_comfort'] = (
        row['1 Bedroom Apartment in City Centre'] / 30
    )

    # --- ENTERTAINMENT ---
    results['entertainment_budget'] = (
        row['Cinema Ticket (International Release)'] / 7
    )
    results['entertainment_mid'] = (
        row['Cinema Ticket (International Release)'] / 4 +
        row['Domestic Draft Beer (0.5 Liter)']
    )
    results['entertainment_comfort'] = (
        row['Cinema Ticket (International Release)'] / 3 +
        row['Bottle of Wine (Mid-Range)'] / 3
    )

    # --- EXTRAS ---
    results['extras_budget'] = (
        row['Bottled Water (1.5 Liter)']
    )
    results['extras_mid'] = (
        row['Bottled Water (1.5 Liter)'] +
        row['Cappuccino (Regular Size)']
    )
    results['extras_comfort'] = (
        row['Bottled Water (1.5 Liter)'] +
        row['Cappuccino (Regular Size)'] * 2 +
        row['Soft Drink (Coca-Cola or Pepsi, 0.33 Liter Bottle)']
    )

    return pd.Series(results)

df_tiers = df_usd[['city']].copy()
df_tiers = pd.concat([df_tiers, df_usd.apply(calculate_daily_costs, axis=1)], axis=1)

df_tiers.head(5)
```
