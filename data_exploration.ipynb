```python
import pandas as pd

df = pd.read_csv("numbeo_raw.csv")

print(df.shape)
print(df.dtypes)
df.head(10)
```

```python
import re

def parse_price(val):
    if pd.isna(val):
        return None
    # Remove currency symbols, letters, spaces — keep digits, dot, comma
    cleaned = re.sub(r'[^\d,\.]', '', str(val))
    # Remove commas used as thousands separators
    cleaned = cleaned.replace(',', '')
    try:
        return float(cleaned)
    except:
        return None

# Apply to all columns except 'city' and the already-numeric mortgage column
price_cols = [c for c in df.columns if c not in ['city', 'Annual Mortgage Interest Rate (20-Year Fixed, in %)']]

df_clean = df.copy()
df_clean[price_cols] = df[price_cols].map(parse_price)

print(df_clean.dtypes)
df_clean.head(3)
```

```python
missing = df_clean.isnull().sum()
missing_pct = (missing / len(df_clean) * 100).round(1)

summary = pd.DataFrame({'missing_count': missing, 'missing_pct': missing_pct})
summary[summary.missing_count > 0].sort_values('missing_pct', ascending=False)
```

```python
drop_cols = [
    'Domestic Non-Alcoholic Beer (0.5 Liter Bottle)',
    'Domestic Draft Non-Alcoholic Beer (0.5 Liter)',
    'Imported Non-Alcoholic Beer (0.33 Liter Bottle)',
    'Bottle of Non-Alcoholic Wine (Mid-Range)',
    'Buffalo Round or Equivalent Back Leg Red Meat (1 kg)'
]

df_clean = df_clean.drop(columns=drop_cols)
print(df_clean.shape)
```

```python
price_cols = [c for c in df_clean.columns if c not in ['city', 'Annual Mortgage Interest Rate (20-Year Fixed, in %)']]

df_clean[price_cols] = df_clean[price_cols].fillna(df_clean[price_cols].median())

# Also fill mortgage rate with its median
df_clean['Annual Mortgage Interest Rate (20-Year Fixed, in %)'] = df_clean['Annual Mortgage Interest Rate (20-Year Fixed, in %)'].fillna(
    df_clean['Annual Mortgage Interest Rate (20-Year Fixed, in %)'].median()
)

print(df_clean.isnull().sum().sum())  # Should be 0
```
